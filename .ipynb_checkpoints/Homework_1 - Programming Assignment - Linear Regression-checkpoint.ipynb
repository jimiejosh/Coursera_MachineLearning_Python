```python

xfiledir = os.path.join(os.getcwd(),'machine-learning-ex1/ex1/')

# print(os.listdir(xfiledir))

# Read data from file as pandas dataframe
dfdata = pd.read_csv(os.path.join(xfiledir,'ex1data1.txt'), header=None, names=['pop','profit'])


city_pop = dfdata['pop']                 # Features X = x1. Single feature
truck_profit = dfdata['profit']             # Profit y 
nos_ex = truck_profit.count()        # Nos of training examples (m)


colors = np.random.rand(nos_ex)
areas = np.pi * (15 * np.random.rand(nos_ex)) * 2

plt.figure(1)
plt.axis([4, 25, -5, 30])
plt.xlabel("Population of city in 10,000s")
plt.ylabel("Profit in $10,000s")
plt.scatter(city_pop, truck_profit, c=colors, s=areas)

plt.figure(2)
plt.plot(dfdata)

plt.figure(3)
plt.plot(dfdata.cumsum())


```
